```python
import cv2
import tensorflow
import keras
from PIL import Image
```

ອ່ານ File ທີ່ໃຊ້ໃນການຈັບໃບໜ້າ

```python
face_cascade = 'haarcascade_frontalface_default.xml'
```

ທຳການເປີດກ້ອງທີ່ຕິດກັບຄວມພິວເຕີທີ່ເກັບໄວ້ໃນຕົວແປ webcam

```python
webcam = cv2.VideoCapture(0)
```

ຄຳສັ່ງ read() ເປັນຄຳສັ່ງທີ່ໃຊ້ອ່ານພາບຈາກ webcam ແຕ່ຂໍ້ມູນທີ່ໄດ້ຈາກ webcam ແມ່ນຂໍ້ມູນປະເພດ BGR

```python
success, image_bgr = webcam.read()
```

ການສົ່ງຂໍ້ມູນຮູບພາບ ຫຼາຍໆຮູບພາບຕິດຕໍ່ກັນ ພວກເຮົາຕ້ອງໃຊ້ while loop ເພື່ອສົ່ງຄຳສັ່ງຂອງພາບຈາກ webcam ເພື່ອສະແດງເປັນວິດີໂອ

```python
while True:
    success, image_bgr = webcam.read()
    if success:
        
        image_bw = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
        
        face_classifier = cv2.CascadeClassifier(face_cascade)
        faces = face_classifier.detectMultiScale(image_bw)
        print(f'There are {len(faces)} faces found.')
        for face in faces:
            x, y, w, h = face
            cv2.rectangle(image_bgr, (x, y), (x+w, y+h), (0,255,0), 2)
        cv2.imshow("Face Detection", image_bgr)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
webcam.release()
cv2.destroyAllWindows()
```

ສ້າງຂໍ້ມູບຂອງຄົນທີ່ໃສ່ Mask ແລະ ບໍ່ໃສ່ mask ເພື່ອນຳໄປ train model

```python
count = 0
while True:
    success, image_bgr = webcam.read()
    if success:
        image_org = image_bgr.copy()
        image_bw = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
        
        face_classifier = cv2.CascadeClassifier(face_cascade)
        faces = face_classifier.detectMultiScale(image_bw)
        print(f'There are {len(faces)} faces found.')
        for face in faces:
            x, y, w, h = face
            cv2.rectangle(image_bgr, (x, y), (x+w, y+h), (0,255,0), 2)
            cv2.imwrite(f'mask/mask_{count}.jpg', image_org[y:y+h, x:x+w])
            count+=1
        cv2.imshow("Face Detection", image_bgr)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
webcam.release()
cv2.destroyAllWindows()
```

### ທຳຄວາມເຂົ້າໃຈ Medel ເບື້ອງຕົ້ນ

```python
from keras.models import load_model
from PIL import Image, ImageOps
import numpy as np

# Load the model
model = load_model('keras_model.h5')

# Create the array of the right shape to feed into the keras model
# The 'length' or number of images you can put into the array is
# determined by the first position in the shape tuple, in this case 1.
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
# Replace this with the path to your image
image = Image.open('mask_3.jpg')
#resize the image to a 224x224 with the same strategy as in TM2:
#resizing the image to be at least 224x224 and then cropping from the center
size = (224, 224)
image = ImageOps.fit(image, size, Image.ANTIALIAS)

#turn the image into a numpy array
image_array = np.asarray(image)
# Normalize the image
normalized_image_array = (image_array.astype(np.float32) / 127.0) - 1
# Load the image into the array
data[0] = normalized_image_array

# run the inference
prediction = model.predict(data)
print(prediction)
```

### ນຳເອົາ Model ມາປະກອບກັບ webcam ແລະ Mask Detection

 ຂໍ້ມູນທີ່ຈາກ webcam ເປັນຂໍ້ມູນປະເພດ object array ເຊີ່ງຈະຕ້ອງແປງເປັນ Image object ໃນຮູບແບບຂອງ object ທີ່ໄດ້ຈາກການເປີດໄຟສຈາກຮູບ ແລະ ຕ້ອງແປງສີຈາກສີ BGR ເປັນ RGB
 ຫຼັງຈາກທີ່ເຮົາແປງຂໍ້ມູນຂອງ Webcame ແລະປັບໃຫ້ເຂົ້າກັບຂໍ້ມູນທີ່ໃຊ້ຮ່ວມກັບ Model ທີ່ Train ມາແລ້ວ ເຮົາຈະລວມຄຳສັ່ງເຂົ້າກັນ ແລະ ສະແດງຜົນເປັນ real-time

```python
import cv2
import tensorflow
from keras.models import load_model
from PIL import Image, ImageOps
import numpy as np

webcam = cv2.VideoCapture(0)
face_cascade = 'haarcascade_frontalface_default.xml'
face_classifier = cv2.CascadeClassifier(face_cascade)
model = load_model('keras_model.h5')
size = (224, 224)
while True:
    success, image_bgr = webcam.read()
    if success:
        
        image_bw = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
        image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
        
        faces = face_classifier.detectMultiScale(image_bw)
        for face in faces:
            x, y, w, h = face
            cface_rgb = Image.fromarray(image_rgb[y:y+h,x:x+w])
        
            data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)

            image = cface_rgb

            image = ImageOps.fit(image, size, Image.ANTIALIAS)

            image_array = np.asarray(image)

            normalized_image_array = (image_array.astype(np.float32) / 127.0) - 1

            data[0] = normalized_image_array

            prediction = model.predict(data)
            print(prediction)
            if prediction[0][0] < prediction[0][1]:
                cv2.putText(image_bgr, 'MASKED', (x, y-7), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0),2)
                cv2.rectangle(image_bgr, (x, y), (x+w, y+h), (0,255,0), 2)
            else:
                cv2.putText(image_bgr, 'NON_MASKED', (x, y-7), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255),2)
                cv2.rectangle(image_bgr, (x, y), (x+w, y+h), (0,0,255), 2)
        cv2.imshow("Face Detection", image_bgr)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
webcam.release()
cv2.destroyAllWindows()
```
